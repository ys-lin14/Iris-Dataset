```python
#splits data into 80-20 training and testing sets and returns a list containing them
def split(df):
    train_set = df.values.tolist()
    
    for x in range(len(train_set)):
        train_set[x].append(0) #add space foreuclidean distances
    
    test_set = []
    train_test = []
    
    test_size = int(len(train_set) * 0.2)
    for y in range(test_size):
        test_set.append(train_set.pop(random.randint(0, len(train_set) - 1)))
    
    for z in range(len(test_set)):
        test_set[z].append("prediction") #adding for prediction labels
        
    train_test.append(train_set)
    train_test.append(test_set)
                    
    return train_test

#computes euclidean distance in terms of sepal length and width given flowers p and q
def euclidean_distance(p, q):
    distance = math.sqrt((p[0] - q[0]) ** 2 + (p[1] - q[1]) ** 2)
    return distance

#returns a list containing the 5 nearest neighbours of p
def nearest_neighbours(p):
    nn = []
    
    for x in range(len(train_set)):
        train_set[x][5] = euclidean_distance(train_set[x], p)
        nn.append(train_set[x])
        
    nn.sort(key = lambda y: y[5])
    return nn[:6]

#classifies flower p based on its 5 nearest neighbours
def classify(nn, p):
    labels = ["setosa", "versicolor", "virginica"]
    label_count = [0, 0, 0]
    
    for x in range(5):
        for y in nn[x]:
            if (y in labels):
                label_count[labels.index(y)] += 1
           
    p[6] = labels[label_count.index(max(label_count))]
    
    

def knn(df):    
    train_test = split(iris_df)

    train_set = train_test[0]
    test_set = train_test[1]

    for x in range(len(test_set)):
        nn = nearest_neighbours(test_set[x])
        classify(nn, test_set[x])
        
    for y in range(len(test_set)):
        test_set[y].pop(5)
        
    return test_set
    
iris_test = knn(iris_df)
iris_test_df = pd.DataFrame(iris_test, columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", 
                                       "species", "prediction"])

iris_test_scatter = sns.relplot(x = "sepal_length", y = "sepal_width", hue = "species", data = iris_test_df)
iris_test_scatter.set(title = "Sepal Width and Length of Irises - Expected Predictions", 
                 xlabel = "Sepal Length (cm)",  
                 ylabel = "Sepal Width (cm)")
plt.show()

iris_test_scatter = sns.relplot(x = "sepal_length", y = "sepal_width", hue = "prediction", data = iris_test_df)
iris_test_scatter.set(title = "Sepal Width and Length of Irises - Actual Predictions", 
                 xlabel = "Sepal Length (cm)",  
                 ylabel = "Sepal Width (cm)")
plt.show()
```
