```python
#basic data visualization of the iris dataset Fisher (1936)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

sns.set_style("ticks") #dark(grid), white(grid), ticks

iris_df = pd.read_csv("iris.csv") #load csv data
```

```python
#scatterplots of sepal and petal length/width
iris_sepal_scatter = sns.relplot(x = "sepal_length", y = "sepal_width", hue = "species", data = iris_df)
iris_sepal_scatter.set(title = "Sepal Width and Length of Irises", 
                 xlabel = "Sepal Length (cm)",  
                 ylabel = "Sepal Width (cm)")
plt.show()

iris_petal_scatter = sns.relplot(x = "petal_length", y = "petal_width", hue = "species", data = iris_df)
iris_petal_scatter.set(title = "Petal Width and Length of Irises", 
                 xlabel = "Petal Length (cm)",  
                 ylabel = "Petal Width (cm)")
plt.show()
```

```python
#vertical boxplots of sepal length/width
sepal_length_box = sns.boxplot(x = "species", y = "sepal_length", data = iris_df)
sepal_length_box.set(title = "Sepal Length of Irises", 
                 xlabel = "Species",
                 ylabel = "Sepal Length (cm)")
plt.show()

sepal_width_box = sns.boxplot(x = "species", y = "sepal_width", data = iris_df)
sepal_width_box.set(title = "Sepal Width of Irises", 
                 xlabel = "Species",
                 ylabel = "Sepal Width (cm)")
plt.show()
```

```python
#horizontal boxplots of petal length/width
petal_length_box = sns.boxplot(x = "petal_length", y = "species", data = iris_df)
petal_length_box.set(title = "Petal Length of Irises", 
                 xlabel = "Pepal Length (cm)",
                 ylabel = "Species")
plt.show()

petal_width_box = sns.boxplot(x = "petal_width", y = "species", data = iris_df)
petal_width_box.set(title = "Petal Width of Irises", 
                 xlabel = "Petal Width (cm)",
                 ylabel = "Species")
plt.show()
```
